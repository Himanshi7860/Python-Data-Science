```python
import numpy as np
import matplotlib.pyplot as plt
```

# charts and graphs
Types of graph
- Bar chart : used to compare the values of different cartegories of data
- Line chart : used to show trends over time 
- Area chart : used to show trends over tiome,but can also be used to show the difference betweeen two values 
- Pie chart : used to show the proportion of each category of data
- Histogram : usede to show the distribution of a numerical variable
- Scatter plot : used to show the relationship between two numerical variable

```python
rainfall = np.array([399 , 555, 464, 587, 559])
years = np.array([2019, 2020, 2021, 2022, 2023])
```

```python
#line plot
plt.plot(rainfall)
```

```python
plt.plot(years, rainfall)
```

```python
#configuring graph
plt.plot(years, rainfall)
plt.title("Rainfall per year", fontdict={'fontsize': 15, 'color':'green'})
plt.xlabel("years")
plt.ylabel("Rainfall (cm)")
plt.xticks(years)
plt.hlines(y=500, xmin=2019, xmax=2023, color='red', linestyles='--')
plt.show()
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.figure(figsize=(10, 6), facecolor='#efefef')
x = np.array(['Rohan', 'Rahul', 'Raj', 'Ramesh', 'Rakesh'])
y = np.array([60, 80, 90, 20, 50])
y2 = np.array([50, 70, 80, 10, 40])
bars = plt.bar(x, y, 
        color=['orange'],
        alpha=.5, edgecolor='black',
        linewidth=1, hatch='//', width=.5)
plt.plot(x, y, color='blue', marker='d', linestyle='--', alpha=.1, linewidth=2)
bars[2].set_color('red')
plt.title("Data")
plt.ylim(0, 120)
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, 
             bar.get_height()+2, 
             f"${bar.get_height()}k",
             fontsize=10, color='red',
             ha='center')
plt.grid(axis='y', alpha=.5)
plt.show()
```

```python
plt.figure(figsize=(5, 5), facecolor='white', dpi=100)
plt.pie(
    y, 
    labels = x,
    explode = [.02]*x.size,
    autopct='%.1f%%',
    pctdistance=.75,
    counterclock=True,
    startangle=90,
    wedgeprops={'width':.5, 'edgecolor':'gray'}, # donut chart
    textprops={'fontsize':10}
)
plt.show()
```
