```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'Canada.xlsx'
df = pd.read_excel(path, skiprows=20, sheet_name=1, skipfooter=2)
df
```

```python
pd.set_option('display.max_column', None)
```

```python
#adding a column
```

```python
years = list(range(1980,2013))
print(years)
```

```python
df[years].sum(axis=1)
```

```python
df['Total'] = df[years].sum(axis=1)
df
```

```python
print(df.columns.tolist())
```

```python
#type,coverage,AREA,REG,DEV
df.drop(columns=['Type', 'Coverage', 'AREA', 'DEV'], inplace=True)
df
```

```python
df.rename(mapper={
    'OdName': 'Country',
    'AreaName': 'Continent',
    'RegName': 'Region',
    'DevName': 'Status'
}, axis=1, inplace=True )
df
```

```python
#all countries with 'A'
df[df.Country.str.startswith("A")]
```

```python
#all countries ending with i using quary
df.query("Country.str.endswith('i')")
```

```python
#all countries with immigration less than 10000
df.query("Total < 10000")
```

```python

df.set_index('Country', inplace=True)
```

```python
df.loc['Japan']
```

```python
df.loc['Japan', years].plot(figsize=(15,5), style='--o')
```

```python
df.loc['India', years].plot(figsize=(15,5), style='--o')
```

```python
df.loc[['India','China', 'Japan'], years].T.plot(figsize=(15,5), kind='area',
                                                 stacked=False, alpha=.3)
```

```python
plt.figure(facecolor="white")
colors = ['#0f8','#f80','#0f5','#0f6','#0f7','#0333']
df.groupby('Continent')['Total'].sum().plot(kind='pie', autopct='%.1f%%',
                                            figsize=(8,8), wedgeprops={
                                                'width':.6,
                                                'edgecolor':'black',
                                            })
plt.title(f"Half of the immigrants are from Asia", fontsize=28)
```

```python
sns.displot(data=df.query('Total>25000'), x='Total', kde=False, col='Status')
```
