```python
import seaborn as sns
import numpy as np
```

```python
x  = np.random.normal(size=100)
sns.histplot(x, kde=True)
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.set_theme(style='darkgrid')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', aspect=1.5, height=4)
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex',
            palette=['yellow', 'green'], edgecolor='black', aspect=2)
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time",
)
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip",
    size="total_bill", aspect=2,
    sizes=(20, 200), alpha=.5, edgecolor="blue",
)
```

```python
dowjones = sns.load_dataset("dowjones")
dowjones
```

```python
sns.relplot(data=dowjones, x="Date", y="Price", kind='line', aspect=2)
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", kind='line', aspect=2)
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", kind='line', aspect=2,
            hue='region')
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", kind='line', aspect=1,
            col='region')
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", kind='line', aspect=1,
            row='region')
```
